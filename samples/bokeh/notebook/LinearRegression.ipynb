```python
from ipywidgets import interact
import numpy as np
import random
import math
import time
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure


output_notebook()
```

```python
x_start = -30
x_end = 70
a = 5
b = 0.6
data_count = 100
random_range = 20

x = np.arange(x_start-20, x_end+20, 1)
y = a + b * x

a_learn = 0.2
b_learn = -5
y_learn = a_learn + b_learn * x

data_x = np.random.randint(x_start, x_end, [data_count])
data_y = a + b * data_x + (0.5 - np.random.random([data_count]))*random_range
```

```python
p = figure(title="线性回归", plot_height=600, plot_width=1000, y_range=[-40, 100], x_range=[-60, 100])
# 数据集
p.circle(data_x, data_y, fill_color="blue", line_color="blue", size=8, fill_alpha=0.6)
# 原始直线
p.line(x, y, legend="真实", color="#ff0000", line_width=2)
# 正在学习的直线
line_learning = p.line(x, y_learn, legend="学习", color="#ff0000", line_width=2, line_dash="4 4")
# 正在学习的点
point_learning = p.circle([0], [0], fill_color="red", line_color="red", size=8)

show(p, notebook_handle=True)
```

```python
leraning_rate = 0.0001
lost_arr = []
stop_time = 1

for _ in range(1000):
    #随机选择一个数
    rand_ind = random.randint(0, data_x.shape[0]-1)
    xx = data_x[rand_ind]
    yy = data_y[rand_ind]
    
    point_learning.data_source.data['x'] = [xx]
    point_learning.data_source.data['y'] = [yy]
    push_notebook()
    
    time.sleep(stop_time)
    
    #计算cost
    y_predict = a_learn + b_learn * xx
    cost = math.pow((yy - y_predict), 2)
    lost_arr.append(cost)
    #计算梯度
    a_grad = y_predict - yy
    b_grad = (y_predict - yy) * xx
    #更新参数
    a_learn = a_learn - a_grad * leraning_rate
    b_learn = b_learn - b_grad * leraning_rate
    
    line_learning.data_source.data['y'] = a_learn + b_learn * x
    p.title.text = 'loss: %.5f' % cost
    push_notebook()
    
    time.sleep(stop_time)
    
#     print cost, a_learn, b_learn
```
